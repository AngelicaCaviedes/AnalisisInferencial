# Regresión Lineal Simple


# 0. Importar librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# 1. Exploración de los datos

```python
datos = pd.read_csv("https://raw.githubusercontent.com/AngelicaCaviedes/AnalisisInferencial/main/clase4/Ejemplo1/ingreso.csv")
datos
```

<BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR>
    

```python
plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color="blue")
plt.show()
```

<BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR><BR>

# 2. Creación del Modelo de Regresión Lineal Simple

```python
regresion = linear_model.LinearRegression()

horas = datos["horas"].values.reshape((-1, 1))
# Se debe vectorizar los datos puesto que podemos tener multiples variables que me hagan
horas
```

```python
modelo = regresion.fit(horas, datos["ingreso"])
# Ajustar una linea a los datos datos

print("Intersección (a)", modelo.intercept_)
print("Pendiente (b)", modelo.coef_)
```

## Y= a + bX + e

```python
entrada = [[39.5], [40], [43], [44]]
modelo.predict(entrada)
```

```python
plt.scatter(entrada, modelo.predict(entrada), color="red")
plt.plot(entrada, modelo.predict(entrada), color="black")

plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color="pink")
plt.show()
```
