# Regresión lineal simple

Para este ejercicio queremos predecir el valor de un hogar dependiendo del área de la misma.

## 0. Cargar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# lectura de datos en Python
train = pd.read_csv('https://raw.githubusercontent.com/AngelicaCaviedes/AnalisisInferencial/main/clase4/Ejemplo%202/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/AngelicaCaviedes/AnalisisInferencial/main/clase4/Ejemplo%202/test.csv')
```

## 1. Exploración de los datos

```python
train.head(3)
```

```python
train.columns
```

```python
train[['GrLivArea','SalePrice']].head()
```

```python
train.plot.scatter(x='GrLivArea',y='SalePrice')
plt.show()
```

```python
X_train = np.array(train['GrLivArea']).reshape((-1, 1))
Y_train = np.array(train['SalePrice'])
```

```python
# creando modelo
model = LinearRegression()
model.fit(X_train, Y_train)
print(f"intercepto (a): {model.intercept_}")
print(f"pendiente (b): {model.coef_}")
```

```python
entrada = [[1710], [2700], [1110], [1550]]
model.predict(entrada)
```

```python
plt.scatter(entrada, model.predict(entrada), color="red")
plt.plot(entrada, model.predict(entrada), color="black")

plt.ylabel("área")
plt.xlabel("Precio de la casa")
plt.scatter(train['GrLivArea'], train['SalePrice'], color="pink")
plt.show()
```
