# Prueba de hipotesis sobre la correlación

La prueba de correlación de Spearman se utiliza cuando se desea evaluar la relación monotónica (no necesariamente lineal) entre dos variables. Monotónica significa que, a medida que aumenta (o disminuye) el valor de una variable, el otro también tiende a aumentar (o disminuir), pero no necesariamente a un ritmo constante.

```python
# Seleccionar dos variables para la prueba de correlación de Spearman
variable2 = casas['SalePrice']
variable1 = casas['GrLivArea']
```

```python
# Calcular la correlación de Spearman y el valor p
correlation_coefficient, p_value = spearmanr(variable1, variable2)
# Imprimir los resultados
print(f'Correlación de Spearman: {correlation_coefficient}')
print(f'Valor p: {p_value}')
```

```python
alpha = 0.05
if p_value < alpha:
    print('La correlación es estadísticamente significativa (rechazamos la hipótesis nula)')
else:
    print('La correlación no es estadísticamente significativa (no hay evidencia suficiente para rechazar la hipótesis nula)')
```

```python
selected_columns = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'YearBuilt']

# Crear un nuevo DataFrame con solo las columnas seleccionadas
selected_df = casas[selected_columns]

# Calcular la matriz de correlación de Spearman
correlation_matrix, _ = spearmanr(selected_df)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5, xticklabels=selected_columns, yticklabels=selected_columns)
plt.title('Correlación de Spearman entre Variables Seleccionadas')
plt.show()
```
